## Some sneaky Linear Algebra: linearising the fit to exponential sums

```python
piControl_baseline_N = MIPs['CMIP']['piControl']['rtmt']['rtmt'][:].squeeze()[-400:].mean()
piControl_baseline_T = MIPs['CMIP']['piControl']['tas']['tas'][:].squeeze()[-400:].mean()
piControl_baseline_CO2 = MIPs['CMIP']['piControl']['co2mass']['co2mass'][:].squeeze()[-400:].mean()

T = MIPs['CMIP']['abrupt-4xCO2']['tas']['tas'][:].squeeze() - piControl_baseline_T
N = MIPs['CMIP']['abrupt-4xCO2']['rtmt']['rtmt'][:].squeeze() - piControl_baseline_N
```

```python
## A function that combines the traditional LinRegr with sp.optimize to calculate the three time constants and coefficients from the abrupt-4x experiment

LR_results = sp.stats.linregress(T,N)

F = LR_results.intercept
T_eq = -LR_results.intercept/LR_results.slope

def fit_tcs(x,T_out=False):
    
    q = np.zeros((1,3))
    q[0,:2] = x[:2]
    q[0,2] = T_eq/F - q.sum()
    
    tau = np.zeros((1,3))
    tau[0,:] = x[2:]
    
    IR_model = np.sum(q*F*(1-np.exp(-np.arange(151)[:,np.newaxis]/tau)),axis=1)
    
    if T_out:
        return IR_model
    
    return np.sum((np.concatenate((np.array([0]),T)) - IR_model)**2)

fit_result = sp.optimize.minimize(fit_tcs,x0=[0.5,0.5,1,15,300],method='Powell')

q = np.zeros(3)
q[:2] = fit_result.x[:2]
q[2] = T_eq/F - q.sum()

tau = np.zeros(3)
tau[:] = fit_result.x[2:]

plt.plot(np.concatenate((np.array([0]),T)))
plt.plot(fit_tcs(fit_result.x,True))
plt.text(0.95,0.05,'q: '+' '.join([str(round(x,2)) for x in q])+'\n'+'tau: '+' '.join([str(round(x,2)) for x in tau]),transform=plt.gca().transAxes,va='bottom',ha='right')

print(fit_result.x)
```

## Tuning Forcing parameters for CO2

```python
## Calculate the required forcing in the 1pct experiment from the parameter fits:

T = MIPs['CMIP']['1pctCO2']['tas']['tas'][:].squeeze() - piControl_baseline_T
N = MIPs['CMIP']['1pctCO2']['rtmt']['rtmt'][:].squeeze() - piControl_baseline_N

F = N-LR_results.slope*T

CO2_ppm = (MIPs['CMIP']['1pctCO2']['co2mass'].variables['co2mass'][:].flatten()) / (5.15e18*44.01/28.97) * 1e6

CO2_PI_conc = piControl_baseline_CO2/ (5.15e18*44.01/28.97) * 1e6

def fit_forc_UKESM(x,forc_out=False):
    
    RF_diagnosed = step_forcing(CO2_ppm,CO2_PI_conc,x)
    
    if forc_out:
        return RF_diagnosed
    
    return np.sum((RF_diagnosed - F)**2)

forc_result = sp.optimize.minimize(fit_forc_UKESM,np.array([5.35,0,0]),method='Nelder-mead')

plt.plot(F)
plt.plot(fit_forc_UKESM(forc_result.x,True))
```

```python
def tune_r0_rC_rT(x):
    
    curr_params = default_gas_forcing_params().copy()
    
    curr_params.loc['f1':'f3',('default','CO2')] = forc_result.x
    curr_params.loc['PI_conc',('default','CO2')] = piControl_baseline_CO2_esm/ (5.15e18*44.01/28.97) * 1e6
    
    beta = curr_params.loc['rC',('default','CO2')] / curr_params.loc['rT',('default','CO2')]
    curr_params.loc['r0',('default','CO2')] = x[0]
    curr_params.loc['rC',('default','CO2')] = x[1]
    curr_params.loc['rT',('default','CO2')] = x[1]/beta
    
    therm_params = pd.concat([pd.DataFrame(index=['d','q'],columns = [1,2,3])],keys=['default'],axis=1).apply(pd.to_numeric)
    
    therm_params.loc['d'] = tau
    therm_params.loc['q'] = q
    
    model_output =run_UnFaIR(emissions_in=ssp585_model_emms,forcing_in=ssp585_model_forc,gas_parameters=curr_params,thermal_parameters=therm_params,show_run_info=False)['C'].xs('CO2',level=-1,axis=1).values.flatten()
    
    return np.sum((model_output-CO2_in)**2)

r0_rC = sp.optimize.minimize(tune_r0_rC_rT,[30,0.02],method='Nelder-mead').x
```
