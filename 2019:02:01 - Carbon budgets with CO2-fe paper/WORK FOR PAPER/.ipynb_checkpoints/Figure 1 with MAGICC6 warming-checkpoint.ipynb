```python
# import RCP85 scenario to stitch onto past
RCP85_E = pd.read_csv('../RCP_data/RCP85_EMISSIONS.csv',skiprows=36,index_col=0)
RCP85_RF = pd.read_csv('../RCP_data/RCP85_MIDYEAR_RADFORCING.csv',skiprows=58,index_col=0)

# import RCP85 scenario to stitch onto past
RCP3_E = pd.read_csv('../RCP_data/RCP3PD_EMISSIONS.csv',skiprows=36,index_col=0)
RCP3_RF = pd.read_csv('../RCP_data/RCP3PD_MIDYEAR_RADFORCING.csv',skiprows=58,index_col=0)
```

```python
# what is the year we run the comparison from (chosen because we need 20 years of non-CO2 forcing data find F_0)
year_val = 2018
non_co2_forcing_average_over = 10

fbar_s = np.array([])
f0_s = np.array([])
f1_s = np.array([])
delta_fs = np.array([])

# Parameter values
full_index = np.arange(1765,2101)
TCR_val = 1.6
ECS_val = 2.75
d_array = np.array([239.0,4.1])
F_2x = 3.74
C_0 = 278.0
H = 100 # size of integration for AGWP calculation
epsilon = (ECS_val - TCR_val)/(TCR_val * d_array[0])
gamma = (1 - np.exp(-epsilon*H))/(epsilon*H)

# magicc like AGWP value
r0 = 32.4 # 33.6
rC = 0.019 # 0.0206
rT = 4.165 # 4.635

base_ems = RCP3_E.FossilCO2.values + RCP3_E.OtherCO2.values
pulse_ems = RCP3_E.FossilCO2.values + RCP3_E.OtherCO2.values
pulse_ems[year_val-1765] += 1. * 12 / 44

other_rf = RCP3_RF.TOTAL_ANTHRO_RF.values - RCP3_RF.CO2_RF.values

C_pulse, T_pulse = fair_scm(emissions=pulse_ems, other_rf=other_rf, tcrecs=np.array([TCR_val,ECS_val]), d=d_array, C_0=C_0, F_2x=F_2x, r0=r0, rC=rC, rT=rT)
C_base, T_base = fair_scm(emissions=base_ems, other_rf=other_rf, tcrecs=np.array([TCR_val,ECS_val]), d=d_array, C_0=C_0, F_2x=F_2x, r0=r0, rC=rC, rT=rT)

RF_net = (F_2x * np.log(C_pulse / C_0) / np.log(2.0)) - (F_2x * np.log(C_base / C_0) / np.log(2.0))
AGWP_CO2 = sp.integrate.simps(RF_net[year_val-1765:year_val+101-1765])
TCRE = (((H / AGWP_CO2) * (F_2x / TCR_val)) * (12 / 44000))**-1
print('TCR: ',TCR_val,'K\nECS: ',ECS_val,'K\ncalculated TCRE: ', TCRE, 'K/TtC')
print(H/AGWP_CO2)
```

```python
# import the datasets from subfolder "15_degree_compatible_scens_IIASA"
co2_ems_df = pyam.IamDataFrame(data='../15_degree_compatible_scens_IIASA/iamc15_snapshot_co2ems.xlsx', encoding='utf-8')
co2_RF_df = pyam.IamDataFrame(data='../15_degree_compatible_scens_IIASA/iamc15_snapshot_co2rf.xlsx', encoding='utf-8')
total_RF_df = pyam.IamDataFrame(data='../15_degree_compatible_scens_IIASA/iamc15_snapshot_totalrf.xlsx', encoding='utf-8')

magicc_df = pyam.IamDataFrame(data='./data/iamc15_fair_magicc_temp_and_forcing.xlsx', encoding='utf-8')

fig1, ax1 = plt.subplots(2,2,figsize=(12,12))

ax1[0,0].set_xlim(2015,2100)
ax1[0,0].set_ylim(-25,50)
ax1[1,0].set_xlim(2015,2100)
ax1[1,0].set_ylim(-0.3,1.3)
ax1[1,1].set_xlim(2015,2100)
ax1[1,1].set_ylim(-0.3,1.3)

ax_nonco2rf = ax1[1,1].twinx()
ax_nonco2rf.set_ylim(-1.0,0.5)

for i in co2_ems_df.models():
    if i in list(magicc_df.filter().models()):
        for j in co2_ems_df.filter(model=i).scenarios():
            if j in list(magicc_df.filter(model=i).scenarios()):
#                 print(i, j)
                # sort, interpolate and plot CO2 ems
                single_co2_ems = co2_ems_df.filter(model=i,scenario=j,variable='Emissions|CO2').as_pandas().value.values
                single_co2_ems_yrs = co2_ems_df.filter(model=i,scenario=j,variable='Emissions|CO2').as_pandas().year.values

                yrs = np.arange(1765,2101)
                co2_ems = np.interp(yrs, single_co2_ems_yrs, single_co2_ems)*12/44000
                co2_ems[:single_co2_ems_yrs[0]-1765+1] = (RCP85_E.FossilCO2.loc[:single_co2_ems_yrs[0]].values + RCP85_E.FossilCO2.loc[:single_co2_ems_yrs[0]].values) * single_co2_ems[0]*12/44000 / (RCP85_E.FossilCO2.loc[single_co2_ems_yrs[0]] + RCP85_E.FossilCO2.loc[single_co2_ems_yrs[0]])
                
                ax1[0,0].plot(yrs, co2_ems*44/12, color='darkblue')
                
                # plot cumulative CO2 ems
                cum_co2_ems = (np.cumsum(co2_ems) - np.sum(co2_ems[:2019-1765]))*44/12000
                ax1[1,0].plot(yrs, cum_co2_ems, color='darkblue')
                
                # sort, interpolate and plot total RF
                single_total_rf = total_RF_df.filter(model=i, scenario=j, variable='Forcing').as_pandas().value.values
                single_total_rf_yrs = total_RF_df.filter(model=i, scenario=j, variable='Forcing').as_pandas().year.values
                    
                total_rf = np.interp(yrs, single_total_rf_yrs[single_total_rf!=0.0], single_total_rf[single_total_rf!=0.0])
                total_rf[:single_total_rf_yrs[0]-1765+1] = RCP85_RF.TOTAL_ANTHRO_RF.loc[:single_total_rf_yrs[0]].values * single_total_rf[0] / RCP85_RF.TOTAL_ANTHRO_RF.loc[single_total_rf_yrs[0]]
                
                # sort, interpolate and plot CO2 RF
                single_co2_rf = co2_RF_df.filter(model=i, scenario=j, variable='Forcing|CO2').as_pandas().value.values
                single_co2_rf_yrs = co2_RF_df.filter(model=i, scenario=j, variable='Forcing|CO2').as_pandas().year.values
                    
                co2_rf = np.interp(yrs, single_co2_rf_yrs[single_co2_rf!=0.0], single_co2_rf[single_co2_rf!=0.0])
                co2_rf[:single_co2_rf_yrs[0]-1765+1] = RCP85_RF.CO2_RF.loc[:single_co2_rf_yrs[0]].values * single_co2_rf[0] / RCP85_RF.CO2_RF.loc[single_co2_rf_yrs[0]]
                
                # derive non-CO2 RF and plot
                nonco2_rf = total_rf - co2_rf
                
                ax_nonco2rf.plot(yrs, nonco2_rf - nonco2_rf[2018-1765], color='darkblue', linestyle=':')
                
                # derive and plot non-CO2 CO2-fe ems
                nonco2_rf_plus = np.append(nonco2_rf, [nonco2_rf[-1], nonco2_rf[-1], nonco2_rf[-1]])
                T_total, E_nonCO2 = fair_scm_emsback(nonco2_rf, other_rf=co2_rf, TCR=1.6, ECS=2.75, d1=4.1, d2=239.0, C_0=278.0)
                
                cum_nonco2_co2fe_ems = (np.cumsum(E_nonCO2) - np.sum(E_nonCO2[:2019-1765]))*44/12000
                ax1[1,1].plot(yrs, cum_nonco2_co2fe_ems, color='darkblue')
                
                
                # find temps from MAGICC and compare to fair derived temps
                single_temp = magicc_df.filter(model=i, scenario=j, variable='AR5 climate diagnostics|Temperature|Global Mean|MAGICC6|Expected value').as_pandas().value.values
                single_temp_yrs = magicc_df.filter(model=i, scenario=j, variable='AR5 climate diagnostics|Temperature|Global Mean|MAGICC6|Expected value').as_pandas().year.values
                
                cum_ems = cum_co2_ems + cum_nonco2_co2fe_ems
                ax1[0,1].plot(cum_ems[single_temp_yrs[0]-1765:] - cum_ems[2018-1765], single_temp - single_temp[single_temp_yrs==2018], color='darkblue', alpha=0.5)

                ax1[0,1].plot(cum_co2_ems[single_temp_yrs[0]-1765:] - cum_co2_ems[2018-1765], single_temp - single_temp[single_temp_yrs==2018], color='darkblue', linestyle=':')
                
```
