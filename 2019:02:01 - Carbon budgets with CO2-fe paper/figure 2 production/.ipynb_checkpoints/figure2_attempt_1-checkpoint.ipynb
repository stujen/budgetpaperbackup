```python
import numpy as np
import scipy as sp
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.mlab as mlab

from scipy.ndimage.filters import gaussian_filter as gfilt

from fair_scm import *
```

```python
rf_200 = np.genfromtxt('./rf_a_reord.txt').reshape((200,267))
rf_best_estimate = np.genfromtxt('./Annualforcings_Mar2014_GHGrevised.txt', skip_header=4).reshape((267,15))

temps_obs = np.genfromtxt('./temp_obs.txt')
```

```python
num = 200

d_vals = np.ones((2,num))
d_vals[0,:] = 409.5
d_vals[1,:] = np.random.lognormal(1.2998, 0.506, num)

RWF_vals = np.random.normal(0.6,0.0915,num)
TCR_vals = np.random.lognormal(0.458, 0.279, num)
ECS_vals = TCR_vals/RWF_vals

h = plt.hist(TCR_vals, bins = 100)
plt.figure()
g = plt.hist(ECS_vals, bins = 100)
plt.figure()
j = plt.hist(d_vals[1,:], bins = 100)
```

```python
fig, ax = plt.subplots(3,1,figsize=(12,17))

count = 0

values_to_run = np.array([])

for j in range(0,200):
    print(j)
    for i in range(0,200):
        C_best, T_best = fair_scm(other_rf=rf_best_estimate[:,14], tcrecs=np.array([TCR_vals[i],ECS_vals[i]]), d=d_vals[:,i])
        T_back, E_back = fair_scm_emsback(rf_best_estimate[:,14], TCR=TCR_vals[i], ECS=ECS_vals[i], d1=d_vals[0,i], d2=d_vals[1,i])

        C_calc, T_calc = fair_scm(other_rf=rf_200[j,:], tcrecs=np.array([TCR_vals[i],ECS_vals[i]]), d=d_vals[:,i])
        T_calc_back, E_calc_back = fair_scm_emsback(rf_200[j,:], TCR=TCR_vals[i], ECS=ECS_vals[i], d1=d_vals[0,i], d2=d_vals[1,i])   
        
        if T_calc[-1] > 0.8 and T_calc[-1] < 1.2:
            ax[0].plot(np.arange(1750,2017), T_calc, color='darkblue', linewidth=1)
            count += 1
            values_to_run = np.append(values_to_run, [i,j])
            
#         else:
#             ax[0].plot(np.arange(1750,2017), T_calc, color='darkblue', linewidth=1)

            ax[1].plot(np.arange(1750,2016), gfilt(E_calc_back[:-1], 3, mode='nearest'), color='darkblue', linewidth=1)
            ax[2].plot(np.cumsum(gfilt(E_calc_back[:-1], 3, mode='nearest')) - np.cumsum(gfilt(E_calc_back[:-1], 3, mode='nearest'))[1870-1850], T_calc[:-1] - np.mean(T_calc[:50]), color='darkblue', linewidth=1)
        
# for k in range(0,count):
#     forcing_timeseries = rf_200[int(values_to_run[k]),:]
#     TCR_value = TCR_vals[int(values_to_run[k+1])]
#     ECS_value = ECS_vals[int(values_to_run[k+1])]
#     d1_value = d_vals[0,int(values_to_run[k+1])]
#     d2_value = d_vals[1,int(values_to_run[k+1])]
    
print(count)
```

```python
TCRE_vals = np.array([])
TCRE_vals_est = np.array([])

H = 100
gamma = 0.86
epsilon = 0.003
AGWP_CO2 = H / 1092.
delta_t = rf_200[0,:].size - 100.

for k in range(0,count):
    if k%100==0:
        print(int(k/100))
    forcing_timeseries = rf_200[int(values_to_run[k]),:]
    TCR_value = TCR_vals[int(values_to_run[k+1])]
    ECS_value = ECS_vals[int(values_to_run[k+1])]
    d1_value = d_vals[0,int(values_to_run[k+1])]
    d2_value = d_vals[1,int(values_to_run[k+1])]
    
    T_calc_back, E_calc_back = fair_scm_emsback(forcing_timeseries, TCR=TCR_value, ECS=ECS_value, d1=d1_value, d2=d2_value)
    
    F_0 = np.mean(forcing_timeseries[81:101])
    F_1 = np.mean(forcing_timeseries[-20:-1])
    F_bar = np.mean(forcing_timeseries[100:-1])
    
    G = 555.0
    
    non_CO2_part = gamma * H * (F_1 - F_0 + (epsilon*delta_t*F_bar)) / (AGWP_CO2*44/12)
    TCRE_val = (T_calc_back[-1] - np.mean(T_calc_back[1850-1750:1901-1750])) / (G + non_CO2_part)
    
    TCRE_vals_est = np.append(TCRE_vals_est, TCRE_val)
    
    TCRE_vals = np.append(TCRE_vals, (T_calc_back[-1] - np.mean(T_calc_back[1850-1750:1901-1750])) / (np.cumsum(E_calc_back)[-1] - np.cumsum(E_calc_back)[1870-1750]))   
    
h = plt.hist(TCRE_vals, bins = 100)
plt.figure()

g = plt.hist(TCRE_vals_est, bins = 100)
plt.show()
```

```python
h = plt.hist(TCRE_vals*1000, bins = 40)

from scipy.stats import norm



mean = np.mean(TCRE_vals*1000)
variance = np.var(TCRE_vals*1000)
sigma = np.sqrt(variance)
x = np.linspace(min(TCRE_vals*1000), max(TCRE_vals*1000),100)
rv = norm(mean,sigma)
plt.plot(x, rv.pdf(x)*950)
plt.figure()


g = plt.hist(TCRE_vals_est*1000, bins = 40)
plt.show()
```
